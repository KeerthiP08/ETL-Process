2.**TRANSFORM**



*   Handling missing values



```python
# --- Electricity_access_percent ---
# Removing empty rows
data_eap_cleaned = pd.read_csv('/content/sample_data/electricity_access_percent.csv',skiprows=4)
display(data_eap_cleaned)
```

```python
# Displaying the cleaned data
display(data_eap_cleaned)
```

```python
# --- rural_population_percent ---
# Removing empty rows
data_rpp_cleaned = pd.read_csv('/content/sample_data/rural_population_percent.csv',skiprows=4)
display(data_rpp_cleaned)
```

```python
# Replacing null values with 0
data_rpp_cleaned = data_rpp_cleaned.fillna(0)
display(data_rpp_cleaned)
```

```python
# --- gdp_data ---
# Removing empty rows
data_gdp_cleaned = pd.read_csv('/content/sample_data/gdp_data.csv',skiprows=4)
display(data_gdp_cleaned)
```

```python
# Replacing null values with 0
data_gdp_cleaned = data_gdp_cleaned.fillna(0)
display(data_gdp_cleaned)
```

```python
# Replacing null values with 0
data_pop_cleaned= data_pop_cleaned.fillna(0)
display(data_pop_cleaned)
```



*   Calculating Additional Metrics(Mean,Sum)



```python
# --- electricity_access_percent ---
# Calculating Mean
data_eap_cleaned['Mean'] = data_eap_cleaned.loc[:, '1960':'2017'].mean(axis=1)
print(data_eap_cleaned['Mean'])
```

```python
# Calculating Sum
data_eap_cleaned['Sum'] = data_eap_cleaned.loc[:, '1960':'2017'].sum(axis=1)
print(data_eap_cleaned['Sum'])
```

```python
# Displaying electricity_access_percent data after cleaning and calculating mean and sum
display(data_eap_cleaned)
```

```python
# --- rural_population_percent ---
# Calculating Average
data_rpp_cleaned['Avg'] = data_rpp_cleaned.loc[:, '1960':'2017'].mean(axis=1)
print(data_rpp_cleaned['Avg'])
```

```python
# Calculating Sum
data_rpp_cleaned['Sum'] = data_rpp_cleaned.loc[:, '1960':'2017'].sum(axis=1)
print(data_rpp_cleaned['Sum'])
```

```python
# Displaying cleaned rural_population_data after calculating average and sum
display(data_rpp_cleaned)
```

```python
# --- gdp data ---
# Calculating Average from 1960 to 2017
data_gdp_cleaned['Avg'] = data_gdp_cleaned.loc[:, '1960':'2017'].mean(axis=1)
print(data_gdp_cleaned['Avg'])
```

```python
# Displaying cleaned gdp data after calculating average and sum
display(data_gdp_cleaned)
```

```python
# --- population_data ---
# Calculating Mean
data_pop_cleaned['Avg'] = data_pop_cleaned.loc[:, '1960':'2017'].mean(axis=1)
print(data_pop_cleaned['Avg'])
```

```python
# Calculating Sum
data_pop_cleaned['Sum'] = data_pop_cleaned.loc[:, '1960':'2017'].sum(axis=1)
print(data_pop_cleaned['Sum'])
```

```python
# country with the highest population for the year 2010 with electricity access and GDP data.
query1 = '''
    SELECT
        p.Country_Name,p."2010" Population,e."2010" Electricity,g."2010" GDP
    FROM
        population_data p,
        Electricity_access e,
        GDP g
    WHERE
    p.Country_Name=e."Country Name" and e."Country Code"=g."Country Code"
    and p."2010"=(select max("2010") from Population_data) and e."2010">0 and g."2010">0
    ;
'''

pd.read_sql(query1, conn)
```

```python
# countries where the access to electricity is less than 20%
query2 = '''
    SELECT *
    FROM Electricity_access
    WHERE Mean<20;
'''
pd.read_sql(query2, conn)
```
